```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pifagorovishtany(n,x0,y0,a,fi,alpha):
# n     
# x0, y0      
# a    
# fi     
# alpha     
    if n==0:
        return
    
    #
    ver1=np.array([x0,y0])
    ver2=ver1+np.array([np.cos(fi)*a,np.sin(fi)*a]) 
    ver3=ver1+np.array([np.sin(-fi)*a,np.cos(fi)*a])
    ver4=ver2+ver3-ver1

    #  
    ax.plot([ver1[0], ver2[0]], [ver1[1], ver2[1]], color="b")
    ax.plot([ver1[0], ver3[0]], [ver1[1], ver3[1]], color="b")
    ax.plot([ver2[0], ver4[0]], [ver2[1], ver4[1]], color="b")
    ax.plot([ver3[0], ver4[0]], [ver3[1], ver4[1]], color="b")

    # 
    tr1=np.cos(alpha)*a
    tr2=np.sin(alpha)*a

    # ,      ""
    ver5=ver3+np.array([tr1*np.cos(fi+alpha),tr1*np.sin(fi+alpha)])

    
    pifagorovishtany(n-1,ver3[0],ver3[1],tr1,fi+alpha,alpha) #  
    pifagorovishtany(n-1,ver5[0],ver5[1],tr2,fi+alpha-np.pi/2,alpha) #   

fig = plt.figure(figsize=(5, 5),dpi=200)
ax=fig.add_subplot(1,1,1)
ax.set_aspect("equal")

pifagorovishtany(7, 1, 1, 3, np.radians(30), np.radians(30))
    
    
```
