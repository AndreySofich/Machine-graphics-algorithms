```python

P1 = np.array([4, 6])  
P2 = np.array([1, 8])  

x1,y1=P1
x2,y2=P2

plt.figure(figsize=(5, 5))

def third_point(t):
   
    return P1 * (1 - t) + P2 * t


def plot_graf(t):
   

    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)
    ax.set_xlabel("X")  
    ax.set_ylabel("Y")  
    ax.set_title("–û—Ç—Ä–µ–∑–æ–∫ P1 P2 –∏ —Ç–æ—á–∫–∞ ")  
    ax.plot([x1, x2], [y1, y2], marker='o', linestyle='-', color='b', label='–û—Ç—Ä–µ–∑–æ–∫')
    
    
    P3 = third_point(t)
    ax.plot(P3[0], P3[1], marker='o', label="P3",color='r')

    ax.text(x1 + 0.1, y1 + 0.01, f'P1({x1}, {y1})', color='b')  
    ax.text(x2 + 0.01, y2 - 0.1, f'P2({x2}, {y2})', color='b')

    ax.legend()  
    ax.grid() 
    plt.show()  
    ax.set_aspect('equal')



slider = widgets.FloatSlider(
    value=0,  
    min=0,    
    max=1,    
    step=0.01,
    description='t',  
)

widgets.interactive(plot_graf, t=slider)


def plot_graf_reverse(t):        
  plot_graf(1 - t)  



widgets.interactive(plot_graf_reverse, t=slider)
```

```python
def rectangle(width, height, ax):

    x_coords = [-width/2, width/2, width/2, -width/2]
    y_coords = [-height/2, -height/2, height/2, height/2]
  
    vertices = list(zip(x_coords, y_coords))
  
    rectangle = patches.Polygon(vertices, closed=True,  facecolor='r', edgecolor='black')
  
    ax.add_patch(rectangle)

    
    ax.set_xlim(-width, width)
    ax.set_ylim(-height, height)
    ax.set_aspect('equal')  
    ax.axhline(0, color='black', linewidth=0.5) 
    ax.axvline(0, color='black', linewidth=0.5) 
    ax.grid(True)
    ax.set_title('–ü—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ ')



fig, ax = plt.subplots()
width = 8
height = 4
rectangle(width, height, ax)
plt.show()
```

```python
def rectangle_1(width, height, ax, transformation_matrix=None):

    x_coords = [-width/2, width/2, width/2, -width/2]
    y_coords = [-height/2, -height/2, height/2, height/2]

    
    vertices_array = np.array([x_coords, y_coords])

    
    if transformation_matrix is not None:
        transformed_vertices = transformation_matrix @ vertices_array
        x_coords = transformed_vertices[0, :]
        y_coords = transformed_vertices[1, :]

    
    vertices = list(zip(x_coords, y_coords))

    
    rectangle = patches.Polygon(vertices, closed=True, facecolor='r', edgecolor='black')

    
    ax.add_patch(rectangle)

   
    min_x = np.min(x_coords)
    max_x = np.max(x_coords)
    min_y = np.min(y_coords)
    max_y = np.max(y_coords)

    ax.set_xlim(min_x - 1, max_x + 1) 
    ax.set_ylim(min_y - 1, max_y + 1)  
    ax.set_aspect('equal')
    ax.axhline(0, color='black', linewidth=0.5) 
    ax.axvline(0, color='black', linewidth=0.5) 
    ax.grid(True)
    ax.set_title('–¢—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫')


fig, ax = plt.subplots()


width = 5
height = 3


transformation_matrix = np.array([[1, -1], [1, 1]])


rectangle_1(width, height, ax, transformation_matrix)


plt.show()
```

3. –í—ã—á–∏—Å–ª–∏—Ç–µ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—É—é –ø–ª–æ—â–∞–¥—å –∏—Å—Ö–æ–¥–Ω–æ–≥–æ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞. –í –∫–∞–∫–æ–º –ø–æ—Ä—è–¥–∫–µ
–Ω—É–∂–Ω–æ –ø–µ—Ä–µ—á–∏—Å–ª–∏—Ç—å –≤–µ—Ä—à–∏–Ω—ã –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞, —á—Ç–æ–±—ã –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –ø–ª–æ—â–∞–¥—å –∏–º–µ–ª–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –∑–Ω–∞–∫?

```python
def oriented_area(vertices):
    n = len(vertices)
    area = 0.0
    for i in range(n):
        j = (i + 1) % n
        area += vertices[i][0] * vertices[j][1]
        area -= vertices[j][0] * vertices[i][1]
    return 0.5 * area

width = 6
height = 4
x_coords = [-width/2, width/2, width/2, -width/2]
y_coords = [-height/2, -height/2, height/2, height/2]
vertices_ccw = list(zip(x_coords, y_coords))


area_ccw = oriented_area(vertices_ccw)
print("–û—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –ø–ª–æ—â–∞–¥—å (–ø—Ä–æ—Ç–∏–≤ —á–∞—Å–æ–≤–æ–π —Å—Ç—Ä–µ–ª–∫–∏):", area_ccw)

vertices_cw = list(zip(x_coords[::-1], y_coords[::-1]))

area_cw = oriented_area(vertices_cw)
print("–û—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –ø–ª–æ—â–∞–¥—å (–ø–æ —á–∞—Å–æ–≤–æ–π —Å—Ç—Ä–µ–ª–∫–µ):", area_cw)


print("–í–µ—Ä—à–∏–Ω—ã –ø—Ä–æ—Ç–∏–≤ —á–∞—Å–æ–≤–æ–π —Å—Ç—Ä–µ–ª–∫–∏: ", vertices_ccw)
print("–í–µ—Ä—à–∏–Ω—ã –ø–æ —á–∞—Å–æ–≤–æ–π —Å—Ç—Ä–µ–ª–∫–µ: ", vertices_cw)
```

```python
# 1. –≠–ª–ª–∏–ø—Å
def plot_ellipse(a, b, marker='o', markersize=3, markevery=1):
    
    t = np.linspace(0, 2 * np.pi, 100)  
    x = a * np.cos(t)
    y = b * np.sin(t)

    fig = plt.figure(figsize=(5, 5))
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(x, y, 'g',  label='–≠–ª–ª–∏–ø—Å') 

    
    ax.plot(x, y, marker=marker, markersize=markersize, markevery=markevery, linestyle='None', color='red', label='–¢–æ—á–∫–∏ —ç–ª–ª–∏–ø—Å–∞')
    
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title("–≠–ª–ª–∏–ø—Å")
    ax.grid(True)
    ax.legend()
    ax.set_xlim(-8,8)
    ax.set_ylim(-8,8)
    
    plt.show()
   
a = 4
b = 2
plot_ellipse(a, b)
```

```python
# 2. –¢—Ä–∞–Ω—Å–ª—è—Ü–∏—è —ç–ª–ª–∏–ø—Å–∞
def plot_translated_ellipse(a, b, dx=1, dy=1, marker='o', markersize=2, markevery=4):
    
    t = np.linspace(0, 2 * np.pi, 100)
    x = a * np.cos(t) + dx
    y = b * np.sin(t) + dy

    fig1 = plt.figure(figsize=(5, 4))
    ax1 = fig1.add_subplot(1, 1, 1)
    ax1.plot(x, y, 'b-', label='–≠–ª–ª–∏–ø—Å')


    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º —Ç–æ—á–∫–∏
    ax1.plot(x, y, marker=marker, markersize=markersize, markevery=markevery, linestyle='None', color='red', label='–¢–æ—á–∫–∏ —ç–ª–ª–∏–ø—Å–∞')

    ax1.set_xlabel("x")
    ax1.set_ylabel("y")
    ax1.set_title("–≠–ª–ª–∏–ø—Å (—Ç—Ä–∞–Ω—Å–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã–π)")
    ax1.grid(True)
    ax1.legend()
    ax.set_xlim(-8,8)
    ax.set_ylim(-8,8)
    
    plt.show()

a = 4
b = 2
plot_translated_ellipse(a, b, dx, dy)
```

```python
# 3. –ì–∏–ø–µ—Ä–±–æ–ª–∞
def plot_hyperbola_branch(a, b, marker='o', markersize=5, markevery=10):
    
    t = np.linspace(0, 2 * np.pi, 100)
    x = a * np.cosh(t)
    y = b * np.sinh(t)

    fig2 = plt.figure(figsize=(8, 6))
    ax2 = fig2.add_subplot(1, 1, 1)
    ax2.plot(x, y, 'g-', label='–ì–∏–ø–µ—Ä–±–æ–ª–∞ (–≤–µ—Ç–≤—å)')

    ax2.plot(x, y, marker=marker, markersize=markersize, markevery=markevery, linestyle='None', color='red', label='–¢–æ—á–∫–∏ –≥–∏–ø–µ—Ä–±–æ–ª—ã')

    ax2.set_xlabel("x")
    ax2.set_ylabel("y")
    ax2.set_title("–ì–∏–ø–µ—Ä–±–æ–ª–∞ (–≤–µ—Ç–≤—å)")
    ax2.grid(True)
    ax2.legend()
    ax2.set_xlim(-10, 10)
    ax2.set_ylim(-10, 10)
    
    plt.show()
    
a = 1
b = 1
plot_hyperbola_branch(a, b)
```
